```python
#Imports
import numpy as np
import random as rnd

def Wash():
    m = 20
    x = []
    y = []
    z = []
    
    #loop to store numbers between 0 and 1 to the array
    for i in range (0,m):
        
        x.append(rnd.uniform(0,1))
        y.append(rnd.uniform(0,1))
        z.append(rnd.uniform(0,1))
    
    xyz = np.asarray((x,y,z))
    Arr = np.transpose(xyz)
    print (Arr,)
    
    
    
    for i in range (0,10):
        
        #Balance with color white
        Arr[:][:][i] = ((Arr[:][:][i]+1)/2)
        
    return Arr

Final = Wash ()

print (Final)
```
