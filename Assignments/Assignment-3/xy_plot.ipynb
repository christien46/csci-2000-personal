```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
from numpy import *

RandomDots = open('xy_points.dat','r')
data1 = RandomDots.readline()
data2 = RandomDots.readline()
RandomDots.close()

#initialize random dots
x = []
y = []



#scan file and collect coordinates of each dot
for i in data1.split():
    i = float(i)
    x.append(float(i))
    
for i in data2.split():
    i = float(i)
    y.append(float(i))
    
x1 = np.asarray(x)
y1 = np.asarray(y)    

xa = x1[x1 < -2]
ya = y1[x1 < -2]

xb = xa[ya >= -5]
yb = ya[ya >= -5]

xc = xb[yb <= 0]
yc = yb[yb <= 0]

xi = np.array(x)
yi = np.array(y)    

a, b = 0, 0


r = 5

y,x = np.ogrid[-a:n-a, -b:n-b]
mask = x*x + y*y <= r*r

array = np.ones((n, n))
array[mask] = 255


plt.figure(figsize=(10,10))  
plt.scatter(x,y, color = 'green')
plt.scatter(x1,y1,color = 'red')
plt.scatter(xc,yc,color = 'blue')
plt.suptitle('Highlighted points (100557844)', fontsize=15)

plt.show()

    
```
