```python
#QUESTION 1:

#First Formula
def rho(n):
    k = 1
    total = 0
    sum = 0
    for i in range(k, n + 1):
        sum = (((-1)**(k+1))/(2*k-1))
        total += sum
        k += 1
    return (4*total)  

rho10 = rho(10)
rho100 = rho(100)
rho1000 = rho(1000)
rho10000 = rho(10000)
rho100000 = rho(100000)
rho1000000 = rho(1000000)

print("First Formula for rho: ")
print ("")
print ('When n = 10, rho_n = ', rho10)
print ('When n = 100, rho_n = ', rho100)
print ('When n = 1000, rho_n = ', rho1000)
print ('When n = 10000, rho_n = ', rho10000)
print ('When n = 100000, rho_n = ', rho100000)
print ('When n = 1000000, rho_n = ', rho1000000)
print ("")



#Second Formula

def tau(n):
    k = 1
    total = 0
    sum = 0
    for i in range(k, n + 1):
        sum = (1/k**2)
        total += sum
        k += 1
    total = total*6
    return(total**(1/2))

tau10 = tau(10)
tau100 = tau(100)
tau1000 = tau(1000)
tau10000 = tau(10000)
tau100000 = tau(100000)
tau1000000 = tau(1000000)


print("Second Formula for tau: ")
print ("")
print ('When n = 10, tau_n = ', tau10)
print ('When n = 100, tau_n = ', tau100)
print ('When n = 1000, tau_n = ', tau1000)
print ('When n = 10000, tau_n = ', tau10000)
print ('When n = 100000, tau_n = ', tau100000)
print ('When n = 1000000, tau_n = ', tau1000000)
print("")



#Third Formula

def mu(n):
    k = 1
    total = 0
    sum = 0
    for i in range(k, n + 1):
        sum = (1/k**4)
        total += sum
        k += 1
    total = total*90
    return(total**(1/4))

mu10 = mu(10)
mu100 = mu(100)
mu1000 = mu(1000)
mu10000 = mu(10000)
mu100000 = mu(100000)
mu1000000 = mu(1000000)


print("Second Formula for mu: ")
print ("")
print ('When n = 10, mu_n = ', mu10)
print ('When n = 100, mu_n = ', mu100)
print ('When n = 1000, mu_n = ', mu1000)
print ('When n = 10000, mu_n = ', mu10000)
print ('When n = 100000, mu_n = ', mu100000)
print ('When n = 1000000, mu_n = ', mu1000000)
print("")





```
