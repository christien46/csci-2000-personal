```python
%matplotlib inline
import matplotlib.pyplot as pyplot

#number of medals in the array
x_medals = [46, 38, 29, 24, 13, 11, 11, 8, 8, 7, 107]

#string values of the medals
labels_medals = ["USA", "CHN", "GBR", "RUS", "KOR", "GER", "FRA", "ITA", "HUN", "AUS", "OTHER"]

#increasing the size of the plot
pyplot.figure(figsize=(15,15))

#manually adjusting the colours
colours = ['maroon','blue','blueviolet','brown','burlywood','cadetblue','chartreuse','chocolate','coral','cornflowerblue','cornsilk','crimson',
'cyan','darkblue']


pyplot.pie(x_medals, labels=labels_medals, startangle = 90, shadow = True, autopct= '%1.1f%%',colors=colours)
#found out how to display percentage cant find out how to display amounts
#Legend below, uncomment if you wish to use however it blocks part of the graph
#pyplot.legend(loc='upper right', frameon=False)

pyplot.title("2012 Olympic gold medals(100557844)")
pyplot.show()
```
